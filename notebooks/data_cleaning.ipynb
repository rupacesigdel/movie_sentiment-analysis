```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import re
import html
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('C:/Users/a2z/OneDrive/Desktop/sentiment-analysis-project/movie_sentiment-analysis/data/raw/imdb.csv')
print(f"Original dataset shape: {df.shape}")
```

```python
df = df.dropna()
```

```python
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = html.unescape(text)
    text = re.sub(r"<.*?>", " ", text)
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text)

    return text.strip()
df = pd.read_csv("C:/Users/a2z/OneDrive/Desktop/sentiment-analysis-project/movie_sentiment-analysis/data/raw/imdb.csv")

df["review"] = df["review"].apply(clean_text)

print(df.head())
```

```python
df.to_csv("C:/Users/a2z/OneDrive/Desktop/sentiment-analysis-project/movie_sentiment-analysis/data/processed/cleaned_data.csv", index=False)
print("✅ Cleaned data saved")
```
